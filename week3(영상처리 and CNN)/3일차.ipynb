```python
def mat_access1(mat):
    for i in range(mat.shape[0]):
        for j in range(mat.shape[1]):
            k = mat[i,j]
            mat[i,j] = k *2
            
def mat_access2(mat):
    for i in range(mat.shape[0]):
        for j in range(mat.shape[1]):
            k = mat.item(i,j)
            mat.itemset((i,j), k * 2)
            
mat1 = np.arange(10).reshape(2,5)
mat2 = np.arange(10).reshape(2,5)

print('원소처리 처리전: ', '\n', mat1)
mat_access1(mat1)
print('원소처리 후:', '\n', mat1)
print()
print()
print('원소 처리 전:', '\n', mat2)
mat_access2(mat2)
print()
print('원소처리 후:', '\n', mat2)
```

```python
import cv2
import numpy as np

image1 = np.zeros((50, 512), np.uint8)
image2 = np.zeros((50, 512), np.uint8)
image3 = np.zeros((50, 512), np.uint8)

rows, cols = image1.shape[:2]

for i in range(rows):
    for j in range(cols):
        image2.itemset((i,j),j//2)
        image3.itemset((i,j), j//20 * 10)
        
cv2.imshow('image1', image1)
cv2.imshow('image2', image2)
cv2.imshow('image3', image3)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
image = cv2.imread("images/bright.jpg", cv2.IMREAD_GRAYSCALE)
if image is None:
    raise Exception("영상 파일 읽기 오류")
    
add_cv2 = cv2.add(image,200)
sub_cv2 = cv2.subtract(image, 200)
add_np = np.add(image,200)
sub_np = np.subtract(image, 200)

cv2.imshow("image", image)
cv2.imshow("add_cv2(bright): Saturation", add_cv2)
cv2.imshow("sub_cv2(dark): Saturation", sub_cv2)
cv2.imshow("add_np(bright): Modulo", add_np)
cv2.imshow("sub_np(dark): Modulo", sub_np)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np

image = cv2.imread("images/contrast.jpg", cv2.IMREAD_GRAYSCALE)

if image is None:
    raise Exception("영상 파일 오류")

noimage = np.zeros(image.shape[:2], image.dtype)
print(image.dtype)

dec_cont = cv2.scaleAdd(image, 0.2, noimage)
inc_cont = cv2.scaleAdd(image, 2.0, noimage)

cv2.imshow("image", image)
cv2.imshow("dec_cont", dec_cont)
cv2.imshow("inc_cont", inc_cont)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np
import matplotlib.pylab as plt

img = cv2.imread("images/pixel.jpg", cv2.IMREAD_COLOR)

if img is None:
    raise Exception("영상 오류 ")
    
b,g,r = cv2.split(img)
histo1 = cv2.calcHist([b], [0], None, [256], [0,255])
histo2 = cv2.calcHist([g], [0], None, [256], [0,255])
histo3 = cv2.calcHist([r], [0], None, [256], [0,255])
# hist_img = draw_histo(histo)


plt.figure(figsize=(6,4))
plt.plot(histo1, color='b')
plt.plot(histo2, color='g')
plt.plot(histo3, color='r')
plt.show()
cv2.imshow('img', img)
# cv2.inshow('hist_img', hist_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
