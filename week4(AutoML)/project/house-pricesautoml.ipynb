```python
reg = setup(data = train_data, 
             target = 'SalePrice',
             remove_multicollinearity=True,
             remove_outliers = True,
             normalize = True,
             silent = True)
```

```python
my_model = create_model("catboost")
```

```python
tunemodel = tune_model(my_model)
```
