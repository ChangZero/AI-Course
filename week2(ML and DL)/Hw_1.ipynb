## 데이터 로드

```python
from sklearn.datasets import fetch_california_housing
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

housing = fetch_california_housing()
print('feature_names:', housing['feature_names'])

X_train_full, X_test, y_train_full, y_test = train_test_split(housing.data, housing.target, random_state=42)
X_train, X_valid, y_train, y_valid = train_test_split(X_train_full, y_train_full, random_state=42)

scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_valid = scaler.transform(X_valid)
X_test = scaler.transform(X_test)
```

```python
import numpy as np
import pandas as pd
import tensorflow as tf
import tensorflow as tf
from tensorflow import keras

np.random.seed(42)
tf.random.set_seed(42)
```

## 함수형 API빌드

```python
input_ = keras.layers.Input(shape=X_train.shape[1:])
hidden1 = keras.layers.Dense(30, activation="relu")(input_)
concat1 = keras.layers.concatenate([input_, hidden1])
hidden2 = keras.layers.Dense(30, activation="relu")(concat1)
hidden3 = keras.layers.Dense(30, activation="relu")(hidden2)
concat2 = keras.layers.concatenate([hidden2, hidden3])
output = keras.layers.Dense(1)(concat2)
model = keras.models.Model(inputs=[input_], outputs=[output])
```

## 모델 요약

```python
model.summary()
```

## 모델 loss와 optimizer 설정

```python
model.compile(loss="mean_squared_error", optimizer=keras.optimizers.SGD(lr=1e-3))
```

## 모델 훈련

```python
history = model.fit(X_train, y_train, epochs=20, validation_data=(X_valid, y_valid))
```

## 성능 평가

```python
mse_test = model.evaluate(X_test, y_test)
print("mse_test", mse_test)
```
